```python
import json
from collections import Counter

# JSON 파일 경로
json_file_path1 = "/content/drive/MyDrive/project/KDT_kiosk/data/est_face_emotion/face_crop/UTK_test_label/UTK_test_label.json"
json_file_path2 =   '/content/drive/MyDrive/project/KDT_kiosk/data/est_face_emotion/face_crop/UTK_train_label/UTK_train_label.json'
# JSON 파일 읽기
with open(json_file_path1, 'r') as file:
    data1 = json.load(file)

with open(json_file_path2, 'r') as file:
    data2 = json.load(file)

# 키에 해당하는 값들 추출
key = 'gender'
# {"image": "54_0_3_20170119211310928.jpg", "gender": "M", "age": 50, "race": "3"}
# [age] 0~116 사이의 정수로 연령을 나타냅니다.
# [gender] 0(남자) 또는 1(여자)입니다.
# [race] 0에서 4까지의 정수로, 백인, 흑인, 아시아인, 인도인 및 기타(히스패닉계, 라틴계, 중동인 등)를 나타냅니다.
labels1 = [item[key] for item in data1]
labels2 = [item[key] for item in data2]

# 각 클래스별 빈도 계산
counts1 = Counter(labels1)
counts2 = Counter(labels2)
total_count = sum(counts1.values()) + sum(counts2.values())

# 결과 출력
print(f"총 개수: {total_count}개")
print(f"클래스 개수: {len(counts1)}")
# print("각 클래스별 개수:")
for i, count in counts1.items():
    print(f"클래스   {i}   :   {counts2[i] + count}개")
    print(f'비율 :            {(counts2[i] + count)/total_count*100:.2f}%')
```

```python
import json
from collections import Counter

# JSON 파일 경로
json_file_path1 = "/content/drive/MyDrive/project/KDT_kiosk/data/est_face_emotion/face_crop/label/labels_v1.json"
json_file_path2 =   '/content/drive/MyDrive/project/KDT_kiosk/data/est_face_emotion/face_crop/test_label/test_labels_v1.json'
json_file_path3 =   '/content/drive/MyDrive/project/KDT_kiosk/data/est_face_emotion/face_crop/val_label/val_labels_v1.json'
# JSON 파일 읽기
with open(json_file_path1, 'r') as file:
    data1 = json.load(file)

with open(json_file_path2, 'r') as file:
    data2 = json.load(file)

with open(json_file_path3, 'r') as file:
    data3 = json.load(file)

# 키에 해당하는 값들 추출
key = 'emotion'
#         "image": "/content/drive/MyDrive/project/KDT_kiosk/data/est_face_emotion/face_crop/test_img/test_face_0.jpg",
#         "gender": "M",
#         "age": 40,
#         "emotion": "panic",
#         "race": "oriental"

labels1 = [item[key] for item in data1]
labels2 = [item[key] for item in data2]
labels3 = [item[key] for item in data3]

# 각 클래스별 빈도 계산
counts1 = Counter(labels1)
counts2 = Counter(labels2)
counts3 = Counter(labels3)
total_count = sum(counts1.values()) + sum(counts2.values()) + sum(counts3.values())

# 결과 출력
print(f"총 개수: {total_count}개")
print(f"클래스 개수: {len(counts1)}")
# print("각 클래스별 개수:")
for i, count in counts1.items():
    print(f"클래스   {i}   :   {counts2[i] + counts3[i] + count}개")
    print(f'비율 :            {(counts2[i] + counts3[i] + count)/total_count*100:.2f}%')

```
