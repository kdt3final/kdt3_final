```python
import pandas as pd
import json
import glob
from tqdm.auto import tqdm
```

# UTK to Json

```python
UTK_train_image_list = glob.glob('/content/drive/MyDrive/kiosk/face_crop/UTK_train/*.jpg')
UTK_test_image_list = glob.glob('/content/drive/MyDrive/kiosk/face_crop/UTK_test/*.jpg')
```

```python
UTK_train_info = []
temp_info = {}
temp_info['image'] = "image"
temp_info['gender'] = "F"
temp_info['age'] = "30"
temp_info['race'] = "Black"

UTK_train_info.append(temp_info)
UTK_train_info.append(temp_info)
UTK_train_info
```

```python
# [0] 나이. [1] 성별: 0 = 남자, 1 = 여자, [2] 인종 : 0 = 백인. 1 = 흑인, 2 = 아시아인, 3 = 인도인, 4 = 나머지

UTK_train_info = []
count = 0
for data_name in tqdm(UTK_train_image_list):
  temp_info = {}
  temp_info['image'] = data_name.split('/')[-1]
  temp_info['gender'] = "F" if data_name.split('/')[-1].split('_')[1] == '1' else "M"
  temp_info['age'] = int( int(data_name.split('/')[-1].split('_')[0]) / 10 ) * 10
  temp_info['race'] = data_name.split('/')[-1].split('_')[2]

  UTK_train_info.append(temp_info)
```

```python
# [0] 나이. [1] 성별: 0 = 남자, 1 = 여자, [2] 인종 : 0 = 백인. 1 = 흑인, 2 = 아시아인, 3 = 인도인, 4 = 나머지

UTK_test_info = []
count = 0
for data_name in tqdm(UTK_test_image_list):
  temp_info = {}
  temp_info['image'] = data_name.split('/')[-1]
  temp_info['gender'] = "F" if data_name.split('/')[-1].split('_')[1] == '1' else "M"
  temp_info['age'] = int( int(data_name.split('/')[-1].split('_')[0]) / 10 ) * 10
  temp_info['race'] = data_name.split('/')[-1].split('_')[2]

  UTK_test_info.append(temp_info)
```

```python
def data_info(data, data_name):
  temp_data_info = {}
  temp_data_info['age'] = data['age'].value_counts()
  temp_data_info['gender'] = data['gender'].value_counts()
  temp_data_info['race'] = data['race'].value_counts()

  print("Dataset :", data_name)
  print("Data_info :", temp_data_info['age'], temp_data_info['gender'],temp_data_info['race'])
  print()
```
